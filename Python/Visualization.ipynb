## Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

## Univariate -> "One Variable"

```python
sns.stripplot(y='tip', data=tips,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.xlabel('tip($)')
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

# Titanic

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/Data-Science/Python/train.csv')
data.head()
```

```python
print(data.dtypes)
```

```python
data.groupby('Survived')['PassengerId'].count()
```

```python
fig ,axs = plt.subplots(figsize=(11,5))
sns.boxplot(x='Survived', y='Age', data=data)
plt.show()
```

```python
fig ,axs = plt.subplots(figsize=(10,5))
sns.barplot(x='Sex', y='Survived', data=data)
```

```python
sns.boxplot(x='Sex',y="Age", hue="Survived", data=data)
```

```python
sns.barplot(x='Pclass', y="Survived", data=data)
```

```python
sns.barplot(x='Pclass', y="Survived", hue="Sex", data=data)
```

```python
sns.barplot(x="SibSp", y="Survived", data=data)
```

```python
sns.barplot(x="Parch", y="Survived", data=data)
```

```python
survived = data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
```

```python
not_survived = data.loc[data['Survived']==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived")
```

```python
sns.pairplot(data.dropna())
```

```python
grid = sns.FacetGrid(data, col='Survived', row='Pclass', height=2.2, aspect=1.5)
grid.map(plt.hist, "Age", alpha= 0.5, bins = 20)
grid.add_legend()
```

```python
sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()

plt.title("Queenstown")
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.xticks([0,1])
plt.show()

plt.title("Southampton")
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.show()
```
